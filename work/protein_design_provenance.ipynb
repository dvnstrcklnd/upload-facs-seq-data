### This notebook performs the following actions:
1. Establishes a session with an Aquarium instance

2. Gets the entire ProvenanceTrace for a specified set of Plans

3. Gets as-executed protocol backtraces for specified Jobs

4. Converts one of the backtraces to Markdown for easier viewing

### 1. Import libraries and establish a session with an Aquarium instance

```python
import provenance_helper
from provenance_helper import *

aq_session = get_session()
```

### 2. Get the entire ProvenanceTrace for a specified set of Plans
This can take several minutes.

```python
plan_ids = [
    32382
]

plans, trace = get_provenance(aq_session, plan_ids)
```

The errors indicate that some Operations were never run (as part of Jobs). In this case they were replaced by new Operations in revised Plans.

### 3. Get *as-executed* protocol backtraces for Challenge and Label Jobs
Get the raw backtrace using the OperationType name, 'Challenge and Label':

```python
challenge_jobs = jobs_for('Challenge and Label', trace)

job = challenge_jobs[0]
job_backtrace = get_backtrace(job.job_id, aq_session)
```

Display part of the raw backtrace is a JSON object:

```python
pprint.pprint(job_backtrace[0:4])
```

### 4. Convert the backtrace to Markdown for easier viewing

```python
md = markdown(job.job_id, job_backtrace)
display(Markdown(md))
```
